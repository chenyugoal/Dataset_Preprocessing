# birdSnap Preprocessing

At present, birdSnap dataset is not fully preprocessed, because there are many expired url which leads to failed download of image. 

Instead, I use only aroung one fourth of it. 

Only those species that contains more than 90 images are selected out and preprocessed and saved as numpy data file.

```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd

from PIL import Image
from PIL import ImageOps
```

```python
path_dataset = 'F:/Programming/Python/NDD/datasets_ORIGINAL/birdsnap/download/images'
path_txt = 'F:/Programming/Python/NDD/datasets_ORIGINAL/birdsnap/species.txt'

path_save_pad = 'F:/Programming/Python/NDD/datasets_padded/birdsnap'
path_save_resize = 'F:/Programming/Python/NDD/datasets_resized/birdsnap'
```

```python
df = pd.read_csv(path_txt, delimiter = "\t")

df.head()
```

```python
df['dir'][0]
```

```python
# select species that meet requirement
minimum = 90 # minimum requirement

selected_species = []

for i in df['dir']:
    #print(path_dataset + '/' + i)
    file = os.listdir(path_dataset + '/' + i)
    #print(len(file))
    if len(file) >= minimum:
        selected_species.append(i)
len(selected_species)
```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True # For error "image file is truncated (45 bytes not processed)"

for i in selected_species:
    #print(path_dataset + '/' + i)
    file = os.listdir(path_dataset + '/' + i)
    for j in file:
        img = Image.open(path_dataset + '/' + i + '/'+ j)
        
        # pad
        w = img.size[0]
        h = img.size[1]
        M = max(h,w)
        
        padded = ImageOps.pad(img, (M,M))
        
        if not os.path.exists(os.path.dirname(path_save_pad+'/'+ i + '/' + j)):
            try:
                os.makedirs(os.path.dirname(path_save_pad+'/'+ i + '/' + j))
            except OSError as exc: # Guard against race condition
                if exc.errno != errno.EEXIST:
                    raise
                    
        padded.convert('RGB').save(path_save_pad+'/'+ i + '/' + j)
        
        w_check = padded.size[0]
        h_check = padded.size[1]
        
        if w_check == h_check:
            pass
        else:
            print('Not square: '+ i + '/' + j)
            break
        
        # resize
        img0 = tf.io.read_file(path_save_pad+'/'+ i + '/' + j)
        img1 = tf.io.decode_jpeg(img0, channels=3)

        img2 = tf.image.resize(img1, [32,32], method='bilinear', preserve_aspect_ratio=True)

        img3 = tf.image.convert_image_dtype(img2/255.0, dtype=tf.uint8)

        out = tf.io.encode_jpeg(img3, format='', quality=100)

        tf.io.write_file(path_save_resize +'/'+ i + '/' + j, out)

        
```

```python
x_all = []
y_all = []
for label,i in enumerate(selected_species):
    file = os.listdir(path_save_resize+'/'+i)
    for j in file:
        #print(path_save_resize+ i + '/'+ j)
        #print(label)
        img = cv2.imread(path_save_resize+'/'+ i + '/'+ j)
        x_all.append(img)
        y_all.append(label)
x_all = np.array(x_all) # (, 32, 32, 3)
y_all = np.array(y_all) # (,)
```
